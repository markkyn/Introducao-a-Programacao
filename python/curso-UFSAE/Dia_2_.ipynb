## Desvios Condicional
 é uma **estrutura de controle** utilizada em programação para executar diferentes blocos de código com base em uma condição ou expressão booleana, permitindo que o programa tome decisões e siga caminhos diferentes de acordo com os resultados obtidos.

### "*Se...,Então", "mas se" e "caso contrário" (if, else, elif)*

#### Se..., Entao *(if)*:
O "if" é uma estrutura de controle que permite executar um bloco de código apenas se uma determinada condição for ***verdadeira***. A sintaxe básica do "if" em Python é a seguinte:

```python
if <condição>:    # Se essa condição for verdadeira 
    # Então esse bloco de codigo é executado
    # ...
```

````python
idade = int(input("Digite sua idade:"))

if idade >= 18:
  print(f'Você é maior de Idade')
  print(f"Voce pode pedir sua carteira de motorista")
````


#### "Caso Contrário" *(else)*
O "else" é uma estrutura de ***controle complementar ao "if"*** e permite definir um bloco de código a ser executado quando todas as condições anteriores forem falsas. A sintaxe básica do "else" em Python é a seguinte:

```python
if <condicao>:   # SE a condição é verdadeira
  # ENTAO o bloco de código
  # ...
else:  
  # CASO CONTRÁRIO executa esse bloco de código
  # ...
```



````python
periodo = int(input("Digite o numero do seu período: "))

if periodo == 1: # Se o periodo da pessoa for 1
  print("Você é Calouro!")
else: # caso contrario
  print("Você não é Calouro.")
````

#### ",mas se" *(elif)*:
O "elif" é uma estrutura de controle que permite **adicionar condições adicionais**, que serão verificadas somente se as condições anteriores forem falsas. A sintaxe básica do "elif" em Python é a seguinte:

````python
periodo = int(input("Digite o numero do seu período: "))
reprovou = False


if periodo == 1: # SE a pessoa for perio 1 , Entao:
  print("Você é Calouro")
  if not reprovou:
    print("Parabens vc vai para o 2° periodo")
  else:
    print("Infelizmente vc reprovou")
elif periodo == 2: # MAS SE ela for 2° periodo
  print("Você não é Calouro, mas também nao é veterano")
else: # CASO CONTRARIO 
  print("Você é Veterano")

# Continue...
````

## Operadores Lógicos no Python
- Negação : ***not*** ou ***!***
- Conjunção : ***and*** ou ***&&***
- Disjunção : ***or*** ou **||**

````python
x = 6
y = 5

sentenca_A = x >= y
sentenca_B = x < y

print(f"Sentença Verdadeira: { sentenca_A }  | Sentença Falsa: { sentenca_B }\n")

# Vamos Começar a manupular o resultado dessas sentenças (True e False)

negacao_verdade = not sentenca_A

print(f"A negação da Sentença A ({sentenca_A}) resulta em { negacao_verdade }\n")

conjuncao = sentenca_A and sentenca_B

print(f'A conjuncao entre "Sentenca A({sentenca_A}) E Sentenca B({sentenca_B})" resulta em { conjuncao }\n')

disjuncao = sentenca_A or sentenca_B

print(f'A disjuncao: "Sentenca A ({sentenca_A}) OU Sentenca B ({sentenca_B})" resulta em { disjuncao }\n')
````

## Desvios Condicionais com Operadores Lógicos

Podemos utilizar operadores lógicos em desvios condicionais para criar condições mais complexas. Alguns operadores lógicos comumente utilizados são:

- `and`: Retorna verdadeiro apenas se todas as condições forem verdadeiras.
- `or`: Retorna verdadeiro se pelo menos uma das condições for verdadeira.
- `not`: Inverte o valor de uma expressão lógica.

Aqui está um exemplo de código em Python que demonstra o uso de operadores lógicos em desvios condicionais:

```python
idade = 20
tem_carteira = True
tem_id = False

if idade >= 18 and tem_carteira:
    print("Você pode dirigir.")
elif idade >= 18 and not tem_carteira:
    print("Você precisa tirar a carteira de motorista.")
elif idade < 18 and tem_id:
    print("Você pode tirar a carteira de motorista quando fizer 18 anos.")
else:
    print("Você não pode dirigir ainda.")



````python
idade = 18
tem_carteira = True
tem_id = True

if idade >= 18 and tem_carteira:
    print("Você pode dirigir.")
elif idade >= 18 and not tem_carteira:
    print("Você precisa tirar a carteira de motorista para poder dirigir")
elif idade < 18 and tem_id:
    print("Você pode tirar a carteira de motorista quando fizer 18 anos.")
else:
    print("Você não pode dirigir ainda.")
````
