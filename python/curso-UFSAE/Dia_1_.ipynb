### Utilização e Manipulação das variáveis.

  * Manipulação:

````python
#@title (Código) Manipulação de Strings 

# Definindo variáveis do tipo String
ola_ufsae = "olá Pilotos!"

boas_vindas = "Boas Vindas"
seu_nome = "Laiza"

# Imprimindo uma String
print(ola_ufsae)

# Concatenação: Faz a união de Strings
print( boas_vindas + " " + seu_nome )

# F-String: É possivel formatar o texto incluindo as variável no meio da string
  # Devemos: incluir o f no inicio da String e utilizar { nome_da_variavel } para incluir-mos o 
print( f"Nós da UFSAE desejeamos {boas_vindas} a {seu_nome}")

# Capitalize: transformar a primeira letra da primeira palavra em maiúscula.
print( 
    ola_ufsae.capitalize() 
)

# Replace: Troca um trecho de Texto por outro.
print( 
    ola_ufsae.replace("Pilotos", "Programadores")
)

# Upper: Transforma todas as letras em maiúsculas.
print( ola_ufsae.upper() )
````

#### Exemplo:
 - Equação do Segundo grau - na Matemática:

$$
  \Delta = b^2 - 4ac
$$

$$
  x = \frac{-b \pm \sqrt{\Delta}}{2a}
$$

$$
  y = ax^2 + bx + c
$$

Lembre que:

$$
  \sqrt[2]{x} = x^\frac{1}{2}
$$

Vamos transformar essa equação em Python:
 

````python
# @title (Código) Equacao do Segundo Grau {vertical-output: true}

# Declaramos as Variáveis
a = 4
b = 5
c = 6

# Delta
delta = (b ** 2) - ( 4 * a * c )

# Formula de Bhaskara
x = -b + ( delta ** (1/2) ) / (2 * a)

# Equação da reta (2° Grau)
y = ( a * (x**2) ) + ( b * x ) + c

print(f"Para a = {a}, b = {b}, c = {c}")
print(f"O valor de Delta é: {delta}")
print(f"O valor de x é: {x}")
print(f"O valor de y é: {y}")
````

````python
# @title (Código) Gráfico da Equação {vertical-output: true}

# Bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Faixa de valores de x
x = np.linspace(-10, 10, 200)

y = a * (x**2) + b * x + c

# Plot do gráfico
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da Equação do Segundo Grau')
plt.grid(True)
plt.show()
````

- Veja o erro (TypeError) a seguir:

````python
cinco_texto   = "5" # O texto 5 não é do tipo Inteiro , mas sim do tipo String
cinco_inteiro =  8  # Variavel do tipo Inteiro

print(idade + 2) # O Interpretador não sabe se o operador + é de soma ou concatenação entre strings

# Irá ocasionar um erro:  TypeError: can only concatenate str (not "int") to str
````

##### Contornando o erro
- Para contornarmos esse erro precisaremos modificar o tipo da nossa variável.
- Cada tipo contém uma Cod Nome pra mostrar pro interpretador qual tipo estamos nos referindo.
- Tipos:
    - String: ***str***
    - Inteiro: ***int***
    - Flutuante: ***float***
    - Booleano: ***bool***

#### Realizando a Conversão
- Para Realizarmos a conversão do tipo basta colocar o tipo que queremos antes da variável, que estará entre parênteses
  - ***int(variavel)***
  - ***str(variavel)***
  - ***float(variavel)***
- Com a função ***type(variavel)*** você pode verificar o tipo da variável

Por exemplo:
``` py

print(f"O tipo da variável cinco_string é: { type(cinco_string) }")

cinco_int = (int) cinco_string # Agora a variavel cinco_int é um INTEIRO

```



````python
# @title (Código) Conversão de tipo

cinco_string = "5"

print(f"O tipo da variável cinco_string é: { type(cinco_string) }")

cinco_int = int(cinco_string) # Agora a variavel cinco_int é um INTEIRO

print(f"O tipo da variável cinco_int é: { type(cinco_int) }")

print(f"É possivel realizar a operação da variável 'cinco_int' com o inteiro 4 -> SOMA: {  cinco_int + 10  }")
````

### Entrada de Dados

Certamente! Abaixo está um exemplo de como você pode usar a função input() em Python para permitir que o usuário insira dados durante a execução do programa. Vou apresentar um tutorial básico usando o Google Colab.
Uso de Entrada de Dados em Python

O Python fornece a função embutida input() para permitir que você solicite e obtenha entrada do usuário durante a execução de um programa. A função input() pausa a execução do programa e aguarda até que o usuário insira algum texto e pressione a tecla "Enter".

Aqui está um exemplo simples que pede ao usuário para digitar seu nome e, em seguida, exibe uma saudação personalizada:

python

nome = input("Digite seu nome: ")
print("Olá, " + nome + "! Bem-vindo(a)!")

    A função input() recebe uma mensagem como argumento, que será exibida para o usuário como uma solicitação.
    O usuário insere o texto desejado e pressiona "Enter".
    O texto digitado pelo usuário é atribuído à variável nome.
    Em seguida, a saudação personalizada é exibida usando o valor armazenado em nome.

Exemplo Completo

Agora, vamos criar um exemplo completo que solicita ao usuário seu nome, idade e cidade, e, em seguida, exibe as informações coletadas:




````python
nome = input("Digite seu nome: ")
idade = input("Digite sua idade: ")
cidade = input("Digite sua cidade: ")
# O resultado dessas 3 variaveis são do tipo String.

print("Nome: " + nome)
print("Idade: " + idade)
print("Cidade: " + cidade)

mais_velho = int(idade) > 6 # Booleano

# Caso você queira manipular a idade do usuário, lembre de transformalo em inteiro
print("Joao tem 16 anos, você é mais velho que Joao? : " + str(mais_velho) )
````


    Solicitamos ao usuário para inserir o nome, a idade e a cidade em três momentos diferentes, usando a função input().
    Cada valor inserido pelo usuário é armazenado nas variáveis nome, idade e cidade, respectivamente.
    Em seguida, exibimos as informações coletadas usando as variáveis armazenadas.

Agora você pode usar a função input() para permitir que o usuário insira dados durante a execução do seu programa Python no Google Colab ou em qualquer outro ambiente!
